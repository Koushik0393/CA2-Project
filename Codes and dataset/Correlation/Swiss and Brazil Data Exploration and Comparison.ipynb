```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
swiss_data = pd.read_excel("swiss-data.xlsx")
brazil_data = pd.read_excel("brazil-data.xlsx")

swiss_data = swiss_data.drop(swiss_data.columns[[0, -2,-1]], axis=1)
brazil_data = brazil_data.drop(brazil_data.columns[[0, -2,-1]], axis=1)
```

```python
fig, axes = plt.subplots(int(len(swiss_data.columns)/2), 2, figsize=(15, 25)) # Create a figure and subplots

for i, column in enumerate(swiss_data.columns):
    swiss_data[column].plot(ax=axes[int(i/2), i%2])
    axes[int(i/2), i%2].set_title(column)
    axes[int(i/2), i%2].set_xlabel(swiss_data.index.name)

#plt.tight_layout()  # Adjust spacing between subplots
plt.subplots_adjust(wspace=0.15, hspace=0.5)
plt.show()
```

```python
fig, axes = plt.subplots(int(len(brazil_data.columns)/2), 2, figsize=(15, 25)) # Create a figure and subplots

for i, column in enumerate(brazil_data.columns):
    brazil_data[column].plot(ax=axes[int(i/2), i%2])
    axes[int(i/2), i%2].set_title(column)
    axes[int(i/2), i%2].set_xlabel(brazil_data.index.name)

#plt.tight_layout()  # Adjust spacing between subplots
plt.subplots_adjust(wspace=0.15, hspace=0.5)
plt.show()
```

```python
fig, axes = plt.subplots(int(len(swiss_data.columns)/2), 2, figsize=(15, 25)) # Create a figure and subplots

for i, column in enumerate(swiss_data.columns):
    #swiss_data[column].plot(ax=axes[int(i/2), i%2], color='red')
    #brazil_data[column].plot(ax=axes[int(i/2), i%2], color='blue')
    ax_swiss = axes[int(i/2), i%2]
    ax_swiss.plot(swiss_data.index, swiss_data[column], color='red', label="Switzerland")
    ax_swiss.set_ylabel('Switzerland', color='red')
    ax_swiss.tick_params(axis='y', labelcolor='red')


    ax_brazil = ax_swiss.twinx()
    ax_brazil.plot(brazil_data.index, brazil_data[column], color='blue', label='Brazil')
    ax_brazil.set_ylabel('Brazil', color='blue')
    ax_brazil.tick_params(axis='y', labelcolor='blue')

    ax_swiss.set_title(column)
    ax_swiss.set_xlabel(swiss_data.index.name)
    # Add a legend
    ax_swiss.legend(loc='upper left')
    ax_brazil.legend(loc='upper right')

#plt.tight_layout()  # Adjust spacing between subplots
plt.subplots_adjust(wspace=0.30, hspace=0.5)
plt.show()
```
